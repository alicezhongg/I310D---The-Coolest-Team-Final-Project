```python
# importing pandas to use in code
import pandas as pd
```

```python
# reading excels into various dataframes
df1 = pd.read_excel("/content/week1_cleaned.xlsx")
df2 = pd.read_excel("/content/week29_cleaned.xlsx")
df3 = pd.read_excel("/content/week45_cleaned.xlsx")
df4 = pd.read_excel("/content/food1_cycle03.xlsx")
```

```python
# renaming dataframe columns
df1.rename(columns={
    'Select Characteristics': 'Select Characteristics',
    'Unnamed: 1': '1_total',
    'Unnamed: 2': '1_enough_and_all_food',
    'Unnamed: 3': '1_enough_and_some_food',
    'Unnamed: 4': '1_sometimes_not_enough',
    'Unnamed: 5': '1_often_not_enough',
    'Unnamed: 6': '1_did_no_report'
}, inplace=True)

df2.rename(columns={
    'Select Characteristics': 'Select Characteristics',
    'Unnamed: 1': '2_total',
    'Unnamed: 2': '2_enough_and_all_food',
    'Unnamed: 3': '2_enough_and_some_food',
    'Unnamed: 4': '2_sometimes_not_enough',
    'Unnamed: 5': '2_often_not_enough',
    'Unnamed: 6': '2_did_no_report'
}, inplace=True)

df3.rename(columns={
    'Select Characteristica': 'Select Characteristics',
    'Unnamed: 1': '3_total',
    'Unnamed: 2': '3_enough_and_all_food',
    'Unnamed: 3': '3_enough_and_some_food',
    'Unnamed: 4': '3_sometimes_not_enough',
    'Unnamed: 5': '3_often_not_enough',
    'Unnamed: 6': '3_did_no_report'
}, inplace=True)

df4.rename(columns={
    'Unnamed: 0': 'Select Characteristics',
    'Unnamed: 1': 'total',
    'Unnamed: 2': 'enough_and_all_food',
    'Unnamed: 3': 'enough_and_some_food',
    'Unnamed: 4': 'sometimes_not_enough',
    'Unnamed: 5': 'often_not_enough',
    'Unnamed: 6': 'did_no_report'
}, inplace=True)

# dropping characteristic columns so that theres no overlap
df2nohead = df2.drop(columns='Select Characteristics')
df3nohead = df3.drop(columns='Select Characteristics')
```

```python
df_spending = pd.concat([df2.iloc[59:68].reset_index(drop=True), df3nohead.iloc[74:-1].reset_index(drop=True)], axis=1)
df_spending["total"] = df_spending["2_total"] + df_spending["3_total"]

df_spending.style
```

```python
# copying AGE characteristics from previous dfs to new dataframe
df_age = pd.concat([df1.iloc[1:6].reset_index(drop=True), df2nohead.iloc[1:6].reset_index(drop=True), df3nohead.iloc[1:6].reset_index(drop=True)], axis=1)
df_age["total"] = df_age["1_total"] + df_age["2_total"] + df_age["3_total"]

df_age.style
```

```python
# copying EDUCATION characteristics from df1 to new dataframe
df_education = pd.concat([df1.iloc[13:17].reset_index(drop=True), df2nohead.iloc[13:17].reset_index(drop=True), df3nohead.iloc[13:17].reset_index(drop=True)], axis=1)
df_education["total"] = df_education["1_total"] + df_education["2_total"] + df_education["3_total"]

df_education.style
```

```python
# copying INCOME characteristics from previous dfs to new dataframe
df_income = pd.concat([df1.iloc[48:57].reset_index(drop=True), df2nohead.iloc[50:59].reset_index(drop=True), df3nohead.iloc[45:54].reset_index(drop=True)], axis=1)
df_income["total"] = df_income["1_total"] + df_income["2_total"] + df_income["3_total"]

df_income.style
```

```python
# copying UNEMPLOYMENT characteristics from previous dfs to new dataframe
df_unemployment = pd.concat([df1.iloc[30:43].reset_index(drop=True), df2nohead.iloc[37:50].reset_index(drop=True), df3nohead.iloc[32:45].reset_index(drop=True)], axis=1)
df_unemployment["total"] = df_unemployment["1_total"] + df_unemployment["2_total"] + df_unemployment["3_total"]

df_unemployment.style
```

```python
# copying ANXIETY characteristics from previous dfs to new dataframe
df_anxiety = pd.concat([df1.iloc[63:68].reset_index(drop=True), df2nohead.iloc[71:76].reset_index(drop=True), df3nohead.iloc[69:73].reset_index(drop=True)], axis=1)
df_anxiety["total"] = df_anxiety["1_total"] + df_anxiety["2_total"] + df_anxiety["3_total"]

df_anxiety.style
```

```python
characteristics = [[],[],[],[],[], []]
percentages2020 = [[],[],[],[],[], []]
percentages2021 = [[],[],[],[],[], []]
percentages2022 = [[],[],[],[],[], []]

dataframes = [df_age, df_unemployment, df_income, df_education, df_anxiety, df_spending]
count = 0
j = 0
i = 0
for dataframe in dataframes:
  count += 1
  for index, row in dataframe.iterrows():
    characteristics[i].append(row['Select Characteristics'])
    percentages2020[i].append([])
    percentages2021[i].append([])
    percentages2022[i].append([])
    if not count == 6:
      for column in dataframe.columns[2:7]:
        percentages2020[i][j].append(row[column] / row['1_total'] * 100)
        print(row[column])
      print("fiin")

    if not count == 6:
      for column in dataframe.columns[8:13]:
        percentages2021[i][j].append(row[column] / row['2_total'] * 100)
    else:
      for column in dataframe.columns[2:7]:
        percentages2021[i][j].append(row[column] / row['2_total'] * 100)

      #print(column)
    if not count == 6:
      for column in dataframe.columns[14:19]:
        percentages2022[i][j].append(row[column] / row['3_total'] * 100)
      #print(column)
    else:
      for column in dataframe.columns[8:13]:
        percentages2022[i][j].append(row[column] / row['3_total'] * 100)
    j+= 1
    print("test")
    print(j)
  i+= 1
  print("done")
  j = 0
print(characteristics)
print(percentages2020[0])
print(percentages2021[0])
print(len(percentages2020[0] + percentages2022[0] + percentages2021[0]))
```
