```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

print("All libraries successfully imported!")
```

```python
week29_data = pd.read_excel("food2_week29.xlsx")
week29_data = week29_data.reset_index(drop=True)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
week29_data
```

```python
# Drop duplicate information
week29_data = week29_data.drop_duplicates()
```

```python
# Drop unecessary information line
week29_data.drop(week29_data.index[1], inplace=True)

# Drop unecessary mortgage and free grocery information (not necessary to the analysis)
week29_data.drop(week29_data.index[149:167], inplace=True)
```

```python
# Drop active duty information
week29_data.drop(week29_data.index[86:117], inplace=True)
```

```python
# Reset index to locate proper indices
week29_data = week29_data.reset_index(drop=True)
week29_data
```

```python
# Drop information regarding respondent sentiments except "Frequency of feeling nervous...."
week29_data.drop(week29_data.index[95:113], inplace=True)
```

```python
# Combine households of 5-7
rows_to_combine = [37, 38, 39]

rows_data = week29_data.iloc[rows_to_combine]

combined_row = ['5-7 people in the household'] + rows_data.iloc[:, 1:].sum().tolist()

week29_data.drop(rows_to_combine, inplace=True)

week29_data = week29_data.reset_index(drop=True)

week29_data.loc[37] = combined_row

week29_data = week29_data.sort_index().reset_index(drop=True)
```

```python
# Drop null values from table
week29_data.fillna('')
```

```python
# Save the data to file and print a confirmation message
week29_data.to_excel("week29_data_cleaned.xlsx", index=False)
print("File saved.")
```
