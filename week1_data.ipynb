```python
# import libraries
import pandas as pd
import numpy as np
print ("Imported all libraries successfully...")
```

```python
# load data
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

week1_data = pd.read_excel("food2a_week1.xlsx")

week1_data.head()
```

```python
# delete duplicates
number_of_duplicates = week1_data.duplicated().sum()
print (f" Number of duplicates before : {number_of_duplicates}")

# Delete duplicate rows
week1_data = week1_data.drop_duplicates()

number_of_duplicates = week1_data.duplicated().sum()
print (f" Number of duplicates after removing : {number_of_duplicates}")
```

```python
# drop unnecessary rows
week1_data.drop(week1_data.index[108:135], inplace=True)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
week1_data
```
